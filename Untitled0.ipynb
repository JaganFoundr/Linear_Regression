```python
import torch

X=torch.tensor([1,2,3,4,5,6,7,8],dtype=torch.float32)
Y=torch.tensor([2,4,6,8,10,12,14,16],dtype=torch.float32)

w=torch.tensor(0.0 ,dtype=torch.float32 ,requires_grad=True)

def forward(x):
  return w*x

def loss(y_actual,y_pred):
  return ((y_pred-y_actual)**2).mean()

x_test=100.0


learning_rate=0.01
nepochs=100


for i in range(nepochs):

  y_prediction=forward(X)

  l=loss(Y,y_prediction)

  l.backward()


  with torch.no_grad():
    w-=learning_rate * w.grad

  w.grad.zero_()

  if (i+1)%10==0:
    print(f"epoch({i+1}) : weights = {w.item():.3f} , loss = {l.item():.3f}")

print("")

print(f"prediction after training : f({x_test}) = {forward(x_test).item()}")
```

```python
import torch
import torch.nn as nn

X=torch.tensor([[1],[2],[3],[4],[5],[6],[7],[8]],dtype=torch.float32)
Y=torch.tensor([[2],[4],[6],[8],[10],[12],[14],[16]],dtype=torch.float32)

n_samples,n_features=X.shape

print(f"n_samples = {n_samples} , n_features = {n_features}")

x_test = torch.tensor([10],dtype=torch.float32)

class LinearRegression(nn.Module):
  def __init__(self,input,output):
    super(LinearRegression,self).__init__()

    self.linear = nn.Linear(input,output)


  def forward(self,x):
    return self.linear(x)


input_size,output_size=n_features,n_features
model=LinearRegression(input_size,output_size)

print(f"prediction before training: f({x_test.item()}) = {model(x_test).item()}")


lr=0.01
nepochs=100

loss=nn.MSELoss()
optimizer=torch.optim.SGD(model.parameters(),lr=lr)

for i in range(nepochs):

  y_prediction=model(X)

  l=loss(Y,y_prediction)

  l.backward()

  optimizer.step()

  optimizer.zero_grad()

  if (i+1)%10==0:
    w,b=model.parameters()

    print(f"epochs {i+1} : weights: {w[0][0].item()} and losses: {l.item()}")

print(f"prediction before training: f({x_test.item()}) = {model(x_test).item()}")
```

```python
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

device=torch.device('cuda' if torch.cuda.is_available else 'cpu')

input_size=784
hidden=500
output_size=10
nepochs=3
batch_size=100
lr=0.001

train_dataset=torchvision.datasets.MNIST(root='./data',download=True,train=True,transform=transforms.ToTensor())
test_dataset=torchvision.datasets.MNIST(root='./data',train=False,transform=transforms.ToTensor())

train_loader=torch.utils.data.DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
test_loader=torch.utils.data.DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=False)

iteration = iter(test_loader)

example_data,example_target=next(iteration)

for i in range(6):
  plt.subplot(2,3,i+1)
  plt.imshow(example_data[i][0],cmap='gray')

plt.show()

class Neuralnetwork(nn.Module):
  def __init__(self,input_size,hidden,output_size):
    super(Neuralnetwork,self).__init__()

    self.input_layer=nn.Linear(input_size,hidden)
    self.acti_fn=nn.ReLU()
    self.output_layer=nn.Linear(hidden,output_size)

  def forward(self,x):
    output=self.input_layer(x)
    output=self.acti_fn(output)
    output=self.output_layer(output)
    return output

model=Neuralnetwork(input_size,hidden,output_size).to(device)

loss_fn=nn.CrossEntropyLoss()

optimizer=torch.optim.Adam(model.parameters(),lr=lr)

for epoch in range(nepochs):
  for i, (images,labels) in enumerate(train_loader):
    images=images.reshape(-1,28*28).to(device)
    labels=labels.to(device)

    prediction=model(images)

    loss=loss_fn(prediction,labels)

    loss.backward()

    optimizer.step()
    optimizer.zero_grad()

    if (i+1)%100==0:
      print(f"epochs: {epoch+1} and loss: {loss}")




with torch.no_grad():
  n_correct=0
  n_samples=len(test_loader.dataset)
  for images,labels in test_loader:
    images=images.reshape(-1,28*28).to(device)
    labels=labels.to(device)

    output=model(images)

    _,prediction=torch.max(output,1)

    n_correct+=(prediction==labels).sum().item()

  acc=n_correct/n_samples
  print(f"Model accuracy is {acc*100}%")
```

```python
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

device=torch.device('cuda' if torch.cuda.is_available else 'cpu')

nepochs=10
batch_size=32
lr=0.001

transform=transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))
    ])

train_dataset=torchvision.datasets.CIFAR10(root='./data',download=True,train=True,transform=transform)
test_dataset=torchvision.datasets.CIFAR10(root='./data',train=False,download=True,transform=transform)

train_loader=torch.utils.data.DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
test_loader=torch.utils.data.DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=False)

classes={'plane','car','bird','cat','deer','dog','frog','horse','ship','truck'}


def imshow(imgs):
  imgs=imgs/2+0.5
  npimgs=imgs.numpy()
  plt.imshow(np.transpose(npimgs,(1,2,0)))
  plt.show

iteration=iter(train_loader)
images,labels=next(iteration)
image_grid=torchvision.utils.make_grid(images[0:25],nrow=5)
imshow(image_grid)




class ConvNN(nn.Module):
  def __init__(self):
    super().__init__()
    self.conv1=nn.Conv2d(3,32,3)
    self.MaxPool=nn.MaxPool2d(2,2)
    self.conv2=nn.Conv2d(32,64,3)
    self.conv3=nn.Conv2d(64,64,3)
    self.fc1=nn.Linear(64*4*4,64)
    self.fc2=nn.Linear(64,10)
    self.acti_fn=nn.ReLU()

  def forward(self,x):
    x=self.acti_fn(self.conv1(x))
    x=self.MaxPool(x)
    x=self.acti_fn(self.conv2(x))
    x=self.MaxPool(x)
    x=self.acti_fn(self.conv3(x))
    x=torch.flatten(x,1)
    x=self.acti_fn(self.fc1(x))
    x=self.fc2(x)
    return x

model=ConvNN().to(device)

loss_fn=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(),lr=lr)

for epoch in range(nepochs):
  for i,(images,labels) in enumerate(train_loader):
    images=images.to(device)
    labels=labels.to(device)

    prediction=model(images)

    loss=loss_fn(prediction,labels)

    loss.backward()

    optimizer.step()
    optimizer.zero_grad()

    if (i+1)%1000==0:
      print(f"epochs: {epoch+1} and loss: {loss}")
print("training finished")
PATH='./cnn.pth'
torch.save(model.state_dict(),PATH)
```

```python
loadmodel=ConvNN()
loadmodel.load_state_dict(torch.load(PATH))
loadmodel.to(device)
loadmodel.eval()


with torch.no_grad():
  n_correct_pred=0
  n_samples=len(test_loader.dataset)

  for images,labels in test_loader:
    images=images.to(device)
    labels=labels.to(device)

    output=model(images)

    _,prediction=torch.max(output,1)

    n_correct_pred+=(prediction==labels).sum().item()

  accuracy=n_correct_pred/n_samples
  print(f"accuracy of the model with {n_samples} samples is {100*accuracy}%")

```

```python
# linear regression - import all the necessary libraries
import torch
import numpy as np

#inputs (temperature, rainfall, humidity)
inputs=np.array([[73,67,45],
                 [91,88,64],
                 [87,134,58],
                 [102,43,37],
                 [69,96,70]],dtype='float32')

print('inputs: ',inputs)

print("---------------------------------------------")

#targets (apples, oranges)
targets=np.array([[56,70],
                 [81,101],
                 [119,133],
                 [22,37],
                 [103,119]],dtype='float32')
print('targets: ',targets)

print("---------------------------------------------")

#convert from numpy format to tensor format
inputs=torch.from_numpy(inputs)
print('tensor inputs: ',inputs)

print("---------------------------------------------")

targets=torch.from_numpy(targets)
print('tensor targets: ',targets)

print("---------------------------------------------")

#defining weights and biases
w = torch.randn(2,3,requires_grad=True)
print('random weights: ',w)

print("---------------------------------------------")

b = torch.randn(2, requires_grad=True)
print('random bias: ',b)

print("---------------------------------------------")

#defining the model
def model(x):
  return (x @ w.t() + b)

print("---------------------------------------------")

#defining a loss function
def mse(pred,tar):
  difference=pred-tar
  return torch.sum(difference*difference)/difference.numel()

predictions = model(inputs)
print("prediction: ",predictions)

print("---------------------------------------------")

#training loop
for i in range(1000):

  #making predctions
  predictions = model(inputs)

  # calculating the loss
  loss=mse(predictions,targets)

  print("loss: ",loss)

  #computing the gradients
  loss.backward()

  #updating the weights without tracking the gradients
  with torch.no_grad():
    lr=0.00001
    w-=w.grad * lr
    b-=b.grad * lr
    #emptying the gradient
    w.grad.zero_()
    b.grad.zero_()

print("---------------------------------------------")


#checking the new loss
print('loss: ',loss)

print("---------------------------------------------")

print("new_prediction: ",predictions)





```
